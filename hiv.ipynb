```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from hiv_patient import HIVPatient
```

# simulate unhealthy patient

```python
patient = HIVPatient()
```

```python
s = patient.reset()
dur = 400
nb_steps = int(dur//1e-3)
print(nb_steps)
states = [s]
for i in range(nb_steps):
    s1 = s + patient.der(s,patient.action_set[0])*1e-3
    s = s1
    states.append(s)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 15))
npst = np.array(states)
axs[0,0].plot(npst[:,0])
axs[0,0].set_title("T1")
axs[0,1].plot(npst[:,1])
axs[0,1].set_title("T1*")
axs[0,2].plot(npst[:,2])
axs[0,2].set_title("T2")
axs[1,0].plot(npst[:,3])
axs[1,0].set_title("T2*")
axs[1,1].plot(npst[:,4])
axs[1,1].set_title("V")
axs[1,2].plot(npst[:,5])
axs[1,2].set_title("E")
```

```python
states[-1]
```

```python
print(np.max(npst, axis=0)-np.min(npst, axis=0))
```

Ok, so the "infected, unhealthy" state is stable.

# State bounds

```python
from tqdm import trange

patient = HIVPatient()
trajs = []
trajs_title = []
dur = 800
nb_steps = int(dur//1e-3)
# uninfected
s = patient.reset(mode="uninfected")
states = [s]
for i in trange(nb_steps):
    s1 = s + patient.der(s,patient.action_set[0])*1e-3
    s = s1
    states.append(s)
trajs.append(states)
trajs_title.append("uninfected")

# early infection
s = patient.reset(mode="uninfected")
s[1] = 1e-4
s[3] = 1e-4
s[4] = 1
states = [s]
for i in trange(nb_steps):
    s1 = s + patient.der(s,patient.action_set[0])*1e-3
    s = s1
    states.append(s)
trajs.append(states)
trajs_title.append("early infection")

# infected healthy
s = patient.reset(mode="healthy")
states = [s]
for i in trange(nb_steps):
    s1 = s + patient.der(s,patient.action_set[0])*1e-3
    s = s1
    states.append(s)
trajs.append(states)
trajs_title.append("healthy")

# infected healthy with 25% immunity failure
s = patient.reset(mode="healthy")
s[5] *= .75
states = [s]
for i in trange(nb_steps):
    s1 = s + patient.der(s,patient.action_set[0])*1e-3
    s = s1
    states.append(s)
trajs.append(states)
trajs_title.append("25% immunity fail")
```

```python
nptrajs = np.array(trajs)
print(nptrajs.shape)
```

```python
print(np.min(nptrajs, axis=(0,1)))
print(np.max(nptrajs, axis=(0,1)))
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 15))
var_names = ["T1", "T1*", "T2", "T2*", "V", "E"]

for var_idx, ax in enumerate(fig.axes):
    for traj_idx in range(nptrajs.shape[0]):
        ax.plot(nptrajs[traj_idx,:,var_idx], label=trajs_title[traj_idx])
    ax.set_title(var_names[var_idx])
    ax.legend()
```

```python
bounds = np.zeros((2,6))
bounds[1,0] = 1e6
bounds[1,1] = 5e4
bounds[1,2] = 3200
bounds[1,3] = 80
bounds[1,4] = 2.5e5
bounds[1,5] = 353200

nptrajs_clipped = nptrajs.copy()

for i in range(nptrajs_clipped.shape[2]):
    np.clip(nptrajs_clipped[:,:,i], bounds[0,i], bounds[1,i], out=nptrajs_clipped[:,:,i]);

print(np.min(nptrajs_clipped, axis=(0,1)))
print(np.max(nptrajs_clipped, axis=(0,1)))
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 15))
var_names = ["T1", "T1*", "T2", "T2*", "V", "E"]

for var_index, ax in enumerate(fig.axes):
    for traj in range(nptrajs.shape[0]):
        ax.plot(nptrajs_clipped[traj,:,var_index], label=trajs_title[traj])
    ax.set_title(var_names[var_index])
    ax.legend()
```
